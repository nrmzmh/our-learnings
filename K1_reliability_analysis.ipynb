***

**Reliability** is the probability of a product or system (i.e., equipment, item, or service) operating under specific condition for a certain amount of time *without failure*.

Its key concepts:
+ Probability: the likelihood of success.
+ Function: to cut, heat, or rotate.
+ Specific conditions: e.g., temperature, speed, or pressure.
+ Satisfactory: operate according to a specification, with acceptable compliance.
+ Specific amount of time: minutes, days, months, or cycle.

***

**Failure rate**: the number of times of a product fails in a specified period of time.
$$ \lambda = \frac{Total\ Number\ of\ Failures}{Total\ Operating\ Time} $$

For example:
Assume that there are 5 light bulbs which operated for 1000 hours.

| # Bulb | Total Hours in Operation | Number of Failures |
| :---: | :--- | :---: |
| 1 | Burned out after 422 hours | 1 |
| 2 | Burned out after 744 hours | 1 |
| 3 | Burned out after 803 hours | 1 |
| 4 | Burned out after 678 hours | 1 |
| 5 | Stayed lit for 1000 hours | 0 |
|  | 3647 | 4 |

$$ \lambda = \frac{4\ failures}{3647\ hours} = 0.0011\ failures\ per\ hour $$
 The failure rate is $ 0.0011\ failures\ per\ hour $.

***

**Time to Failure (TTF)**: the duration until a product to fail.
$$ TTF = Total\ Operating\ Time $$

***

**Mean Time to Failure (MTTF)**: the average time for a product to fail (*for non-repairable product*).
E.g., A light bulb is a non-repairable product. When a light bulb burned out, it is replaced.

$$ MTTF = \frac{1}{\lambda} = \frac{Total\ Operating\ Time}{Total\ Number\ of\ Failures} $$

For example:
Assume a product has been running for 1000 hours and experiences 4 failures during that time.

$$ MTTF = \frac{500\ hours}{4\ failures} = 125\ hours $$
 The average time for this product to fail is $ 125\ hours $.

+ MTTF is a useful metric for predicting product reliability and planning for maintenance and repairs.
+ If a product has a MTTF of 125 hours, means that you can  expect it to fail on average every 125 hours.

***

**Mean Time to Repair (MTTR)**: the average time for a product takes to repair (*for repairable product*).
$$ MTTR = \frac{Total\ Maintenance\ Time}{Total\ Numbers\ of\ Repairs} $$

For example:
Assume the maintenanace teams spent 100 hours repairing the failed cars. Over the span of a month, the company experiences 5 car failures.

$$ MTTR = \frac{100\ hours}{5\ failures} = 20\ hours $$
 The average time for maintenance team to repair a failed car is $ 20\ hours $.

+ MTTR is a useful metric for tracking maintenance team effectiveness and identifying areas for improvement.
+ By understanding MTTR helps reduce downtime and costs by improving maintenance procedures and processes.
+ Reducing MTTR, improves product reliability and reduce downtime. This can improve customer satisfaction and profits.

***

**Mean Time Between Failures (MTBF)**: the average time between failures (*for both repairable and non-repairable product*).
E.g., A car engine is a repairable product. When a car engine failes, it is repaired and retured to service.

$$ MTBF = \frac{Total\ Operating\ Time}{Total\ Number\ of\ Failures} $$

For example:
Assume that there are 3 cars.

| # Bulb | Total Hours in Operation | Number of Failures |
| :---: | :--- | :---: |
| 1 | Operated for 200 hours | 3 |
| 2 | Operated for 200 hours | 1 |
| 3 | Operated for 100 hours | 1 |
|  | 500 | 5 |

$$ MTBF = \frac{500\ hours}{5\ failures} = 100\ hours $$
 The average time for car engine to fail is $ 100\ hours $.

+ MTBF is a useful metric for predicting product reliability and planning for maintenance and repairs.
+ If a product has a MTBF of 100 hours, means that you can expect it will fail on average every 100 hours.

***

**Availability**: A measure of the percentage of time a product or system is available for use.


$$ Availability, A = \frac{MTBF}{MTBF\ +\ MTTR} $$

For unavailability, $ U $ is simply $ U = 1 - A $

+ Reliability is the probability of a product operating without failure. A measure of how likely a product functioning correctly.
+ Availability is the probability of a product functioning throughout its lifetime. A measure of how often a product is available for use.

Example use cases for reliability vs availability:

1. The car is 95% reliable, which means it will most likely perform correctly 95% of the time. However, if it is shared equally between two family members, it will be available just 47.5% of the time on average. This is because the car is only available to one family member at a time..

2. The product fails every minute on average, but recovers in just 500ms. This means that the product has a low reliability, but a high availability. The product is only unavailable briefly when it fails. This is because the product is only unavailable for a short period of time when it fails.


Reliability and availability are *not mutually exclusive*. In general, a product with higher reliability generally corresponds to higher availability. However, it is possible for a product to have high availability and low reliability, or vice versa. For example:
+ A product that is frequently being repaired or replaced may have a high availability, but low reliability. This is because the product is quickly brought back into operation after a failure.
+ A reliable product that is only used occasionally, may have low availability. This is because the product is not available for use even when it is functioning correctly.
+ A critical product, like a medical device, may require a very high reliability and availability.
+ A non-critical product, such as a home entertainment product, may not require a high reliability or availability.

***

**Censoring**

**Censored data** is any data for which we do not know the exact event time. Meanwhile, data for which the exact event time is known is referred to as **complete data**.



There are three types of censored data:
1. **Right censored**: data for products that have not yet failed. <br></br>
 <br></br>

2. **Left censored**: data for products that failed before the start of the test. <br></br>
 <br></br>

3. **Interval censored**: when the exact failure time is not known but the lower and upper bounds of an interval surrounding the failure are known. <br></br>
 <br></br>

There are two types of censoring:
1. **Type I**: If a test is stopped after a certain amount of time then the test is *"time-terminated"*.
2. **Type II**: If a test is stopped after a certain number of failures then the test is *"failure-terminated"*.

***


**Reliability Distribution**

 <br></br>

**Age related failure**
+ **Overall Failure Ratio**: *aka "The Bathtub Curve"*, high infant mortality, then low level of random failure, then a wear out zone. Decreasing failure rate, low constant failure rate, then increasing failure rate.
+ **Infant Mortality**: *aka "The Reversed J Curve"* or *"Worst New"*, high infant mortality, then random failure. Quality failures or early stage failure.
+ **Wear-Out**: *aka "The Traditional View"* , random failure then a wear out zone. Wear-out failures or end of life failure.

**Non-age related failure**
+ **Constant Random Failures**: Normal operation. Expected normal life.

## Reliability Analysis Methods

**Life Data Analysis (LDA)**

In LDA, it is assumed that the product being analysed are *non-repairable*. When analysing *non-repairable* component failures, the data is usually in the form of times-to-failure. *Non-repairable products/systems* have independent and identically distributed (i.i.d) lifetimes.

The basic steps involved in life data analysis:
1. Gather life data.
2. Select a lifetime distribution.
3. Estimate the parameters of the distribution.
4. Generate plots and results.

***

**Recurrent Data Analysis (RDA)**

In RDA, it is assumed that the product being analysed are *repairable*. Failures are recurring events in a repairable product, and data is obtained by recording the age of the product at each failure. RDA takes into account the fact that events in recurrent data may be dependent and non-iid. This is because the occurrence of one event can affect the probability or timing of subsequent events.

Repairable systems' failure behaviour depends on their repair history. Traditional LDA methods, like the Weibull distribution, are not suitable because they assume each failure event is identical and independent from the previous one.

***

**Reliability Block Diagrams (RBD)**

A RBD is a graphical representation of a system, showing the functional relationship between its subsystems and components. <br></br>

<br></br>
**Series blocks**: All of the blocks in a series must work for the system to work.
+ If any one block fails, the whole system fails
+ The reliability of the entire series system is the product of the reliability of each individual component:
$$ R_{s} = R_{1}  R_{2}  ...  R_{n} $$

<br></br>
<br></br>
**Parallel blocks**: All blocks in the parallel connection need to fail in order for system failure to occur.
+ If any one block in a parallel system fails, the system will still work as long as at least one other block is working.
+ The redundant component acts as a standby component where it operates if the other component fails. Redundancy in a system occurs when not all elements are necessary for successful operation.
+ The reliability of the entire parallel system can be computed using equation below:
$$ R_{s} = 1 - [(1 - R_{1})  (1 - R_{2})  ...  (1 - R_{n})] $$

The basic steps involved in RBD:
1. Build graphical system model.
2. Define system characteristics.
3. Perform analysis.
4. Review and evaluate.

RBDs are used to analyse the reliability of complex systems, such as aircraft, power plants, and medical devices. 

RBDs are valuable for reliability engineers and system designers. 
+ They analyse system reliability and improve it through redundancy design.
+ RBDs identify failure areas and improve system reliability through redundancy.
+ If a system has a critical component with low reliability, the designer can add a redundant component to improve overall system reliability.

```python
# Identify the hour, minute, and second.
set_hour_minute = set(df_raw["date"].dt.strftime("%H:%M:%S"))

if len(set_hour_minute) == 1:
    frequency_date = "daily"
else:
    frequency_date = "different"

print(f"The frequency of the date is {frequency_date}.")
```

```python
# What is the total device?
total_device = df_raw["device"].nunique()
print(f"There are {total_device} devices in total.")
```

```python
# Keep non-failure device with max date more than "2015-05-01".
max_date = df_raw.groupby("device")["date"].max().reset_index()
df_max_date = pd.merge(df_raw, max_date, on=["device", "date"], how="inner")
df_non_failure = df_max_date[df_max_date["failure"] == 0].copy()
list_non_failure_device = df_non_failure[df_non_failure["date"] >= "2015-05-01"]["device"].tolist()
list_failure_device = df_raw[df_raw["failure"] == 1]["device"].unique().tolist()

df_study = df_raw[df_raw["device"].isin(list_non_failure_device + list_failure_device)]
df_study
```

```python
# Are there any missing value?
print(df_study.isnull().sum())
print(f"There are {df_study.isnull().sum().sum()} missing value.")
```

```python
# Are there any missing date of each device?
df_group = df_study.groupby("device")

list_device_have_missing_date = []
for name, df in df_group:
    missing_date = pd.date_range(
        start=df["date"].min(),
        end=df["date"].max(),
        freq="d").difference(df["date"])

    if len(missing_date) > 0:
        list_device_have_missing_date.append(name)
        # print(f"Device {name} have {len(missing_date)} missing date.")

if len(list_device_have_missing_date) > 0:
    print(f"There are {len(list_device_have_missing_date)} device with missing date.")

    df_study = df_study[~df_study["device"].isin(list_device_have_missing_date)]
    print("Done remove device with missing date.")
```

```python
# Are there any duplicate? One device should have unique timestamp.
total_duplicate = df_study.duplicated(subset=["date", "device"], keep=False).sum()
print(f"There are {total_duplicate} duplicate.")

if total_duplicate > 0:
    display(df_study[df_study.duplicated(subset=["date", "device"], keep=False)])
    df_study = df_study.drop_duplicates()
    df_study = df_study.reset_index(drop=True)
    print("Done drop duplicate.")
```

```python
# Proprocess DataFrame into a non-repairable item.
# Remove device that goes life after failure.
list_failure_device = df_study[df_study["failure"] == 1]["device"].unique().tolist()

max_date = df_study.groupby("device")["date"].max().reset_index()
df_max_date = pd.merge(df_study, max_date, on=["device", "date"], how="inner")
list_failure_max_date = df_max_date[df_max_date["failure"] == 1]["device"].tolist()

list_diff_device = list(set(list_failure_device).difference(list_failure_max_date))
df_diff = df_study[df_study["device"].isin(list_diff_device)]
# df_diff[df_diff["failure"].where(df_diff["failure"] == 1).bfill(limit=1).ffill(limit=2).notnull()]

for ix, row in df_max_date[df_max_date["device"].isin(list_diff_device)].iterrows():
    if row["date"] < df_study["date"].max():
        df_study = df_study[~(df_study["device"] == row["device"])]
```
