***

**Reliability** is the probability of a system or an item (i.e., equipment, product, or service) operating under specific condition for a certain amount of time *without failure*.

Its key concepts:
+ Probability: the likelihood of success.
+ Function: to cut, heat, or rotate.
+ Specific conditions: e.g., temperature, speed, or pressure.
+ Satisfactory: operate according to a specification, with acceptable compliance.
+ Specific amount of time: minutes, days, months, or cycle.

***

**Failure rate**: the number of times of a system fails in a specified period of time.
$$ \lambda = \frac{Total\ Number\ of\ Failures}{Total\ Operating\ Time} $$

For example:
Assume that there are 5 light bulbs which operated for 1000 hours.

| # Bulb | Total Hours in Operation | Number of Failures |
| :---: | :--- | :---: |
| 1 | Burned out after 422 hours | 1 |
| 2 | Burned out after 744 hours | 1 |
| 3 | Burned out after 803 hours | 1 |
| 4 | Burned out after 678 hours | 1 |
| 5 | Stayed lit for 1000 hours | 0 |
|  | 3647 | 4 |

$$ \lambda = \frac{4\ failures}{3647\ hours} = 0.0011\ failures\ per\ hour $$
 The failure rate is $ 0.0011\ failures\ per\ hour $.

***

**Time to Failure (TTF)**: the duration until a system to fail.
$$ TTF = Total\ Operating\ Time $$

***

**Mean Time to Failure (MTTF)**: the average time for a system to fail (*for non-repairable item*).
E.g., A light bulb is a non-repairable item. When a light bulb burned out, it is replaced.

$$ MTTF = \frac{1}{\lambda} = \frac{Total\ Operating\ Time}{Total\ Number\ of\ Failures} $$

For example:
Assume a system has been running for 1000 hours and experiences 4 failures during that time.

$$ MTTF = \frac{500\ hours}{4\ failures} = 125\ hours $$
 The average time for this system to fail is $ 125\ hours $.

+ MTTF is a useful metric for predicting system reliability and planning for maintenance and repairs.
+ If a system has a MTTF of 125 hours, means that you can  expect it to fail on average every 125 hours.

***

**Mean Time to Repair (MTTR)**: the average time for a system takes to repair (*for repairable item*).
$$ MTTR = \frac{Total\ Maintenance\ Time}{Total\ Numbers\ of\ Repairs} $$

For example:
Assume the maintenanace teams spent 100 hours repairing the failed cars. Over the span of a month, the company experiences 5 car failures.

$$ MTTR = \frac{100\ hours}{5\ failures} = 20\ hours $$
 The average time for maintenance team to repair a failed car is $ 20\ hours $.

+ MTTR is a useful metric for tracking maintenance team effectiveness and identifying areas for improvement.
+ By understanding MTTR helps reduce downtime and costs by improving maintenance procedures and processes.
+ Reducing MTTR, improves system reliability and reduce downtime. This can improve customer satisfaction and profits.

***

**Mean Time Between Failures (MTBF)**: the average time between failures (*for both repairable and non-repairable item*).
E.g., A car engine is a repairable item. When a car engine failes, it is repaired and retured to service.

$$ MTBF = \frac{Total\ Operating\ Time}{Total\ Number\ of\ Failures} $$

For example:
Assume that there are 3 cars.

| # Bulb | Total Hours in Operation | Number of Failures |
| :---: | :--- | :---: |
| 1 | Operated for 200 hours | 3 |
| 2 | Operated for 200 hours | 1 |
| 3 | Operated for 100 hours | 1 |
|  | 500 | 5 |

$$ MTBF = \frac{500\ hours}{5\ failures} = 100\ hours $$
 The average time for car engine to fail is $ 100\ hours $.

+ MTBF is a useful metric for predicting system reliability and planning for maintenance and repairs.
+ If a system has a MTBF of 100 hours, means that you can expect it will fail on average every 100 hours.

***

**Availability**: A measure of the percentage of time a system is available for use.


$$ Availability, A = \frac{MTBF}{MTBF\ +\ MTTR} $$

For unavailability, $ U $ is simply $ U = 1 - A $

+ Reliability is the probability of a system operating without failure. A measure of how likely a system functioning correctly.
+ Availability is the probability of a system functioning throughout its lifetime. A measure of how often a system is available for use.

Example use cases for reliability vs availability:

1. The car is 95% reliable, which means it will most likely perform correctly 95% of the time. However, if it is shared equally between two family members, it will be available just 47.5% of the time on average. This is because the car is only available to one family member at a time..

2. The system fails every minute on average, but recovers in just 500ms. This means that the system has a low reliability, but a high availability. The system is only unavailable briefly when it fails. This is because the system is only unavailable for a short period of time when it fails.


Reliability and availability are *not mutually exclusive*. In general, a system with higher reliability generally corresponds to higher availability. However, it is possible for a system to have high availability and low reliability, or vice versa. For example:
+ A system that is frequently being repaired or replaced may have a high availability, but low reliability. This is because the system is quickly brought back into operation after a failure.
+ A reliable system that is only used occasionally, may have low availability. This is because the system is not available for use even when it is functioning correctly.
+ A critical system, like a medical device, may require a very high reliability and availability.
+ A non-critical system, such as a home entertainment system, may not require a high reliability or availability.

***

**Censoring**

**Censored data** is any data for which we do not know the exact event time. Meanwhile, data for which the exact event time is known is referred to as **complete data**.



There are three types of censored data:
1. **Right censored**: data for items that have not yet failed. <br></br>
 <br></br>

2. **Left censored**: data for items that failed before the start of the test. <br></br>
 <br></br>

3. **Interval censored**: when the exact failure time is not known but the lower and upper bounds of an interval surrounding the failure are known. <br></br>
 <br></br>

There are two types of censoring:
1. **Type I**: If a test is stopped after a certain amount of time then the test is *"time-terminated"*.
2. **Type II**: If a test is stopped after a certain number of failures then the test is *"failure-terminated"*.

***


**Reliability Distribution**

 <br></br>

**Age related failure**
+ **Overall Failure Ratio**: *aka "The Bathtub Curve"*, high infant mortality, then low level of random failure, then a wear out zone. Decreasing failure rate, low constant failure rate, then increasing failure rate.
+ **Infant Mortality**: *aka "The Reversed J Curve"* or *"Worst New"*, high infant mortality, then random failure. Quality failures or early stage failure.
+ **Wear-Out**: *aka "The Traditional View"* , random failure then a wear out zone. Wear-out failures or end of life failure.

**Non-age related failure**
+ **Constant Random Failures**: Normal operation. Expected normal life.

### Time-to-Failure (TTF)

```python
list_failure_devices = df[df["failure"] == 1]["device"].unique().tolist()
list_non_failure_devices = list(set(df["device"].unique()).difference(list_failure_devices))

grouped_df = df.groupby("device")
df["time_to_failure"] = grouped_df["failure"].transform(lambda x: range(1, len(x) + 1))
df.loc[df["device"].isin(list_non_failure_devices), "time_to_failure"] = 0
df[df["device"] == list_failure_devices[0]]
```

### Right censored data
There are devices that have not yet failed.

```python
filtered_device = list_failure_devices[:5] + list_non_failure_devices[:5]
filtered_df = df[df["device"].isin(filtered_device)]
max_dates = filtered_df.groupby("device")["date"].max().reset_index()
filtered_df = pd.merge(filtered_df, max_dates, on=["device", "date"], how="inner")
filtered_df["marker"] = filtered_df["failure"].map({0:"o", 1:"x"})
filtered_df["color"] = filtered_df["failure"].map({0:"#008FD5", 1:"#FF2700"})

plt.figure(figsize=(14, 5))
for m in filtered_df["marker"].unique():
    df_marker = filtered_df[filtered_df["marker"] == m]
    g = sns.scatterplot(data=df_marker, x="date", y="device", marker=m, color=df_marker["color"], s=100, zorder=3)

for index, row in filtered_df.iterrows():
    g.hlines(y=row["device"], xmin=df["date"].min(), xmax=row["date"], color="#8B8B8B", linewidth=2)

plt.title("Total of Failure by Month")
plt.show()
```
